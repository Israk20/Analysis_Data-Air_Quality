### Cleaning Data

Disini akan dilakukan pembersihan data agar data dapat digunakan dan reabilitas

We need to fill all of them that has "NaN" values.

Make a function that can use to all of the df, because all of the NaN value in the same columns as the same df. Make an list that has every columns.

```python
#Fungsi untuk mengisi seluruh nilai NaN yang ditemukan di setiap columns pada df yang dipilih
def fill_na_with_most_common(df, columns):
    print(df.isna().sum())
    print()
    print("Here is the result:")
    for n in columns:
        Value_Change = df[n].value_counts()
        df.fillna(value=Value_Change.index[0], inplace=True)
    return df.isna().sum()
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
#Tampilkan seluruh describe dari setiap df
dataframes = {
    "Aotizhongxin_df": Aotizhongxin_df,
    "Changping_df": Changping_df,
    "Dingling_df": Dingling_df,
    "Dongsi_df": Dongsi_df,
    "Guanyuan_df": Guanyuan_df,
    "Gucheng_df": Gucheng_df,
    "Huairou_df": Huairou_df,
    "Nongzhanguan_df": Nongzhanguan_df,
    "Shunyi_df": Shunyi_df,
    "Tiantan_df": Tiantan_df,
    "Wanliu_df": Wanliu_df,
    "Wanshouxigong_df": Wanshouxigong_df
}

for name, df in dataframes.items():
    print(f"Deskripsi {name}:")
    print(df.describe(include="all"))
    print("\n" + "="*50 + "\n")
```

```python
#Menampilkan Temperature diseluruh daerah pada setiap tahun
for name, df in dataframes.items():
  print(f"Station {name}: ")
  print(df.groupby(by="year").agg({
    "No": "nunique",
    "TEMP": ["max", "min", "mean", "std"]
  }))
  print()
```

```python
#Menampilkan Curah Hujan Pada setiap Daerah di china pada setiap tahun
for name, df in dataframes.items():
  print(f"Station {name}: ")
  print(df.groupby(by="year").agg({
    "No": "nunique",
    "RAIN": ["max", "min", "mean", "std"]
  }))
  print()
```

```python
#Menampilkan banyak CO yang dihasilkan pada setiap tahun pada beberapa
total_per_year = pd.DataFrame()
for name, df in dataframes.items():
  print(f"Station {name}: ")
  print(df.groupby(by="year").CO.nunique().sort_values(ascending=False).reset_index().head(10))
  yearly_total = df.groupby('year')['CO'].sum().reset_index()
  yearly_total['station'] = name
  total_per_year = pd.concat([total_per_year, yearly_total])
  print()

total_per_year = total_per_year.groupby('year')['CO'].sum().reset_index()
print("Total banyak CO yang dihasilkan pada setiap tahun: ")
print(total_per_year)
```

Now we need to save all of the data that we got to all_data.csv

```python
all_data = pd.DataFrame()

for name, df in dataframes.items():
    df['station'] = name
    all_data = pd.concat([all_data, df])

all_data.to_csv('all_data.csv', index=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Disini akan menampilkan grafik temperatur selama 2013 hingga 2017 pada setiap daerah

```python
# Menampilkan Temperature diseluruh daerah pada setiap tahun
for name, df in dataframes.items():
    print(f"Station {name}: ")
    grouped_df = df.groupby(by="year").agg({
        "No": "nunique",
        "TEMP": ["max", "min", "mean", "std"]
    })
    print(grouped_df)
    print()

    # menampilkan dalam bentuk line chart
    plt.figure(figsize=(10, 5))
    plt.plot(grouped_df.index, grouped_df[('TEMP', 'mean')], label='Mean Temperature')
    plt.fill_between(grouped_df.index, grouped_df[('TEMP', 'min')], grouped_df[('TEMP', 'max')], color='g', alpha=.1)
    plt.title(f'Temperature Trends for Station {name}')
    plt.xlabel('Year')
    plt.ylabel('Temperature')
    plt.legend()
    plt.show()
```

### Pertanyaan 2:

Menampilkan curah hujan pada setiap daerah di china pada tahun 2013-2017

```python
# Menampilkan Curah Hujan Pada setiap Daerah di china pada setiap tahun
for name, df in dataframes.items():
    print(f"Station {name}: ")
    # Mengubah tipe data 'year' menjadi integer
    df['year'] = df['year'].astype(int)
    grouped_df = df.groupby(by="year").agg({
        "No": "nunique",
        "RAIN": ["max", "min", "mean", "std"]
    })
    print(grouped_df)
    print()

    # Membuat bar chart
    plt.figure(figsize=(10, 5))
    plt.bar(grouped_df.index, grouped_df[('RAIN', 'mean')], label='Mean Rainfall')
    plt.title(f'Rainfall Trends for Station {name}')
    plt.xlabel('Year')
    plt.ylabel('Rainfall')
    plt.legend()
    plt.show()
```

### Pertanyaan 3:

Disini akan menampilkan grafik tentang seberapa banyak CO yang dihasilkan pada setiap tahun untuk seluruh daerah

```python
#Menampilkan banyak CO yang dihasilkan pada setiap tahun pada beberapa
total_per_year = pd.DataFrame()
for name, df in dataframes.items():
  print(f"Station {name}: ")
  print(df.groupby(by="year").CO.nunique().sort_values(ascending=False).reset_index().head(10))
  yearly_total = df.groupby('year')['CO'].sum().reset_index()
  yearly_total['station'] = name
  total_per_year = pd.concat([total_per_year, yearly_total])
  print()

total_per_year = total_per_year.groupby('year')['CO'].sum().reset_index()
print("Total banyak CO yang dihasilkan pada setiap tahun: ")
print(total_per_year)

plt.figure(figsize=(10, 6))
plt.plot(total_per_year['year'], total_per_year['CO'], marker='o')

plt.title('Total Emisi CO per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Emisi CO')

plt.show()
```
